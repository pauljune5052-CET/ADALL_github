```python
import pandas as pd
# Example: Replace this with the raw URL of your GitHub file
github_raw_url = "https://raw.githubusercontent.com/pauljune5052-CET/ADALL_github/refs/heads/main/laptop_prices_2024_sgd_TL.csv"
try:
    df = pd.read_csv(github_raw_url)
    print("Successfully loaded data from GitHub!")
    print(df.info())
    display(df.head())
except Exception as e:
    print(f"Error loading data: {e}")
    print("Please ensure the URL is correct and the file format is compatible with `pd.read_csv`.")
```

```python
#plot each df graph
import matplotlib.pyplot as plt
df.plot(kind='bar')
plt.show()
```

```python
#plot each column in df in numeric using histogram
for column in df.columns:
    df[column].hist()
    plt.title(column)
    plt.show()
```
